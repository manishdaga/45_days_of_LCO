```python
# Generating & ploting sign wave

import matplotlib.pyplot as plt
import numpy as np


Fs = 8000
f = 5
sample = 8000
x = np.arange(sample)
y = np.sin(2 * np.pi * f * x / Fs)
plt.plot(x, y)
plt.xlabel('sample(n)')
plt.ylabel('voltage(V)')
plt.show()
```

```python
# Generating & ploating cosine wave
y = np.cos(2 * np.pi* f * x / Fs )
plt.plot(y)
```

```python
# Generating a tan wave
y = np.tan(2 * np.pi* f * x / Fs )
plt.plot(y)
```

```python
# Generating cot wave
y = 1/np.tan(2 * np.pi* f * x / Fs )
plt.plot(y)
```

```python
# Unit steo Fxn
def uni_step(x):
    if x<0:
        return 0
    else:
        return 1
```

```python
x=np.arange(-10,10)
x
```

```python
y=list(map(uni_step,x))   # map fxn will take the value from x & apply it to the uni_step fxn
y
```

```python
plt.plot(x,y)
```
