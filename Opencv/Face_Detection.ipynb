## Face Detecction in a given image 

```python
import cv2 as cv
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import requests 
import io
```

```python
img=cv.imread('../images/img1.jpg')
plt.imshow(img)
```

```python
img_rgb=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
image_array=np.array(img_rgb)
```

```python
image_array.shape
```

```python
gray=cv.cvtColor(image_array,cv.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
detector=cv.CascadeClassifier("../data/haar_frontal.xml")
```

```python
faces=detector.detectMultiScale(gray,1.3,4)   #1.3 ,5 is standard formate   5 min neigbour b/w two faces pixels
```

```python
faces[0]
```

```python
x,y,w,h=faces[0]
```

```python
cv.rectangle(image_array,(x,y),(x+w,y+h),(255,255,255),2)   # color can be [] & ()
plt.imshow(image_array)
```

```python
# Detect all images
for i in range(len(faces)):
    x,y,w,h=faces[i]
    cv.rectangle(image_array,(x,y),(x+w,y+h),(255,0,0),2)   # color can be [] & ()
    plt.imshow(image_array)
```

## Image Detection using the camera of the system & External Web Cam

```python
# Access the camera of the Laptop 
cam=cv.VideoCapture(0) 
```

```python
status,frame=cam.read() 
```

```python
status   # It show True when it is open otherwise it show False
```

```python
plt.imshow(frame) 
```

```python
frame.shape
```

```python
img_rgb=cv.cvtColor(frame,cv.COLOR_BGR2RGB) 
plt.imshow(img_rgb)
```

```python
frame_gray=cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
```

plt.imshow(frame_rgb)   # Open Cv always read image in BGR formate

```python
detector=cv.CascadeClassifier("../data/haar_frontal.xml")
```

```python
#Detecting Faces
faces=detector.detectMultiScale(frame_gray,1.5,5)  
```

```python
faces
```

```python
x,y,w,h=faces[0]
cv.rectangle(img_rgb,(x,y),(x+w,y+h),(255,255,255),2)   # color can be [] & ()
plt.imshow(img_rgb)
```

```python
cv.imwrite('messigray.png',img_rgb)  #This will save the captured image in current working directory
```
