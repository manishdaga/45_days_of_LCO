```python
#Method 1--> Img combining using hstack
combi1=np.hstack((originalimg,revimg))
plt.imshow(combi1)
```

```python
#Method 2 -->Img combining using vstack
combi2=np.vstack((originalimg,revimg))
plt.imshow(combi2)
```

```python
#Method 2--> image combining using slicing 
wid=originalimg.shape[1]
height=originalimg.shape[0]
canvas=np.ones((height,wid*2,3),dtype='uint8')
canvas[:,:wid]=originalimg
canvas[:,wid:]=revimg
plt.imshow(canvas)
```

```python
originalimg.shape
```

```python
inv=cv.flip(canvas,0)
he=np.vstack((canvas,inv))
plt.imshow(he)
```

```python
h=originalimg.shape[0]
w=originalimg.shape[1]
quadcanvas=np.ones((h*2,w*2,3),dtype='uint8')
quadcanvas[:h,:w]=originalimg
quadcanvas[h:,:w]=cv.flip(originalimg,0)
quadcanvas[:h,w:]=revimg
quadcanvas[h:,w:]=cv.flip(revimg,0)

plt.imshow(quadcanvas)
```
