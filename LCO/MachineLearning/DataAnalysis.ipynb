```python
import requests
import json
temp=[]
city=[]
weather=[]
k=0
```

```python
caps=['Hyderabad',
 'Patna',
 'Agartala',
 'Kohima',
 'Chandigarh',
 'Ranchi',
 'Thiruvananthapuram',
 'Mumbai',
 'Bhubaneswar',
 'Chandigarh',
 'Panaji',
 'Gandhinagar',
 'Dispur',
 'Amaravati',
 'Dehradun',
 'Shimla',
 'Bhopal',
 'Jaipur',
 'Aizawl',
 'Srinagar',
 'Kolkata',
 'Raipur',
 'Chennai',
 'Lucknow',
 'Gangtok',
 'Itanagar',
 'Bangalore',
 'Shillong',
 'Imphal']
```

```python
for i in caps:
    z=requests.get('http://api.openweathermap.org/data/2.5/weather?q='+i+'&APPID=d0483e147fd81229fc4a3162fbfb7e07')
    dic=json.loads(z.content.decode())
    print(type(z.content.decode()))
    if dic['cod']== 200:
        k+=1
        print(k)
        print('City Name:' , dic['name'])
        print('Temperature:' ,dic['main']['temp'])
        print('Weather:', dic['weather'][0]['main'])
        temp.append(dic['main']['temp'])
        city.append(dic['name'])
        weather.append(dic['weather'][0]['main'])

    else:
        print(i,'Spelling is Incorrect')
    
```

```python
import pandas as pd
dt = pd.DataFrame({"city":city,"temp":temp,"weather":weather})
dt
```

## Ploting Lined Graph

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.plot(temp,city,color='green', linestyle='dashed', linewidth = 3, marker ='o' ,markerfacecolor= 'blue', markersize =10)
plt.xlabel('Temperature in Kelvin(K)')
plt.ylabel('State Capitals of Indai')
plt.title('Temperature Variation By State Capitals Of IN')
```

```python
plt.hist(weather,color='skyblue')
plt.title("Weather COunt")
plt.xlabel("Weather Type")
plt.ylabel("Cities")
plt.xticks(rotation=45)    # it will rotate the values on x axis with a certain angle 
plt.yticks(rotation=45)
```

## Seaborn Library

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=dt.city, y=dt.temp,hue = dt.weather)
plt.xticks(rotation = 75)
plt.xlabel('City')
plt.ylabel('Temp')
plt.title('Temperature Of Capitals')
plt.legend(loc='upper right')
```

## Point Plot in Seaborn

```python
f,ax1 = plt.subplots(figsize=(20,10))
sns.pointplot(x=dt.city,y=dt.temp,color='lime')  # Plot Points
plt.xlabel('States',fontsize = 15, color = 'red')
plt.ylabel('Temp',fontsize =15 , color = 'blue')
plt.title('State VS Temperature',fontsize =15 , color = 'blue')
plt.xticks(rotation = 70)
plt.grid() # Put rectangle over the graph to know the exact values
```

## Violinplot --> Used for plot the Density of the data

```python
import seaborn as sns
# Show each ditribution with both violins & points
sns.violinplot(data = sd , inner="points")  # It will plot the column with numeric values present in the DataFrame
plt.show()
sns.violinplot(data = sd , inner="box")  # Here yellow point at the line show the distribution of 75% -25% 
plt.show()
```
